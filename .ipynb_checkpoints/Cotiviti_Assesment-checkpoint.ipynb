## Linear Model Fit

```python
# Function to get 95% confidence intervales
def build_results(x, y):
    
    ## Fit with outliers
    X = sm.add_constant(x) # Add constant
    model = sm.OLS(y, X)
    results = model.fit()
    print(results.summary())
    
    # Get summary data
    stats, data, labels = summary_table(results, alpha=0.05)
    
    # Get confidence intervals (95%)
    ci_low, ci_upp = data[:, 6:8].T
    
    # Plot results
    fig, ax = plt.subplots(figsize = (16,8))
    ax.scatter(x, y, label = "Data")
    ax.plot(x, results.fittedvalues, 'g', label="OLS")
    ax.plot(x, ci_low, 'r')
    ax.plot(x, ci_upp, 'r')
    ax.legend(loc = 'best')
```

### Fit with outliers

```python
## Fit with outliers

# Run function
build_results(x1, y1)
```

### Fit with outliers removed

```python
# Remove outliers based on z score
df_sub = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
print(df_sub.shape) # 3 outliers removed

# Fit
build_results(df_sub['X'], df_sub['Y'])
```

#### Remarks
We notice there is no improvement on the R-squared values with outliers removed. We should proceed with the model containing our full dataset.

## Train Test Split
This section will fit and check the accuracy of our linear model

```python
# Split data (25%)
x2 = x1.values.reshape(-1, 1) # Reshape data for sklearn
x_train, x_test, y_train, y_test = train_test_split(x2, y1, test_size=0.25, random_state=23)

# Fit model
model = LinearRegression().fit(x_train, y_train)

# Check accuracy
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

#### Remarks
We notice the accuracy of our training and test data is 53% and 49% respectively. This is not a great model fit; we likely need more data and a more robust model for a better fit. Also, the high covariance in our data is one likely cause for our low accuracy.
